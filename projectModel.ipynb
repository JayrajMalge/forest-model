```python
datacorr=x.corr()
features=set()
for i in range(len(datacorr.columns)):
    for j in range(i):
        if abs(datacorr.iloc[i,j])>0.80:
            features.add(datacorr.columns[i])

features
```

Deleting highly co-realted columns

```python
x=x.drop(columns=['DC']).reset_index(drop=True)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
import pickle as pic
pic.dump(scaler,open("projectScaler.pkl",'wb'))
```

```python
from sklearn.linear_model import RidgeCV
ridgecv=RidgeCV()
ridgecv.fit(x_train,y_train)
```

```python
y_test
```

```python
y_pred_test=ridgecv.predict(x_test)
```

```python
pic.dump(ridgecv,open("Modelridge.pkl",'wb'))
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred_test)
```
